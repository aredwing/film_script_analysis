The **'fetch'** function requests and receives the contents of a webpage throws and exception if finds a problem, it is used to scrape the web for information.

```python
import operator
import requests
import string
import sys

def fetch(address):
    res = requests.get(address)
    try:
        res.raise_for_status()
    except Exception as exc:
        print('Problem: %s'%(exc))
    return res
```

For the source URL, the variations are in the page number and the page letter, which is the starting letter of the script name. Collect the ranges for all the initial letters.

```python
import bs4
import json

url       = ["http://www.springfieldspringfield.co.uk/movie_scripts.php?order=","&page="]
letters   = string.ascii_uppercase
charNum   = {}

for i in list('0'+letters):
    start = '1'
    main = bs4.BeautifulSoup(fetch(url[0]+i+url[1]+start).text,'lxml')
    temp = main.find_all('a')
    for j in temp[-1:]:
        num = str(j.contents[0]).encode('utf8')
        charNum.update({i:int(num)})


with open('data/charNum.json','w') as f:
    json.dump(charNum,f)
    
print charNum
```

Now that we have the Letter/Length pair, proceed to scrap the links. Because we do not want to be overload the server it, will select randomly one number for each letter, that is, one page randomy visited for per letter. That should serve the purpose of a statistically valid sampling.

The URLs are returned without the root that needs to be added. A dictionary is defined to hold the name of the movie and its link these links contain the droids we are looking for... I mean the scripts.

```python
import pandas as pd
import numpy  as np

prefix = 'http://www.springfieldspringfield.co.uk'
pages  = {}

for i in [0,1,2,3]:
    np.random.seed(i)
    for k in charNum.keys():
        j = np.random.choice(range(1,charNum[k]))
        temp = bs4.BeautifulSoup(fetch(url[0]+k+url[1]+str(j)).text,'lxml')
        aLinks = temp.find_all('a',class_='script-list-item')
        clean = ''
        for link in aLinks:
            pages.update({str(link.contents[0]):prefix+link.get('href')})
        # print k,j
```

To perform the web scrapping specifically from this page. The **'springScrap'** function is defined.

```python
def springScrap(raw):
    result = ' '
    soup = bs4.BeautifulSoup(raw.text,'lxml')
    for e in soup.findAll('br'):
        e.extract()
    for text in soup.find_all('div',class_='scrolling-script-container'):
        result += text.get_text().encode('utf8')
    return result
```

Including or removing the actions may have an impact in the accuracy of the model, test should be made to check which options yields better results.

To remove these actions, the function removeCAPS is defined. After removing them they are saved as well

```python
def removeCAPS(script):
    temp = '' 
    for i in range(len(script)-1):
        if script[i] == ' ' or script[i].islower() or (script[i].isupper() and script[i+1].islower()):
            temp += script[i]
    return ' '.join(temp.split())
```

OMDB allows to extract basic imdb infomation that will be user as meta-data for the regression models, parse from JSON to dictionary.

```python
def omdb(tag=None,name=None):
    if tag:
        url = 'http://www.omdbapi.com/?i='+tag+'&plot=short&r=json'
        raw = fetch(url)
    else:
        url = "http://www.omdbapi.com/?t="+name+"&y=&plot=short&r=json"
        raw = fetch(url)
    result = ''
    for i in raw:
        result += i 
    return json.loads(result)
```

The response from the API contains the information to be used, along with information that should be left out. The function **'flatten'** will create a flat dictionary.

```python
def flatten(orig):
    data = {}
    for c in orig['tree']['children']:
        if 'children' in c:
            for c2 in c['children']:
                if 'children' in c2:
                    for c3 in c2['children']:
                        if 'children' in c3:
                            for c4 in c3['children']:
                                if (c4['category'] == 'personality'):
                                    data[c4['id']] = c4['percentage']
                                    if 'children' not in c3:
                                        if (c3['category'] == 'personality'):
                                                data[c3['id']] = c3['percentage']
    return data
```

## Natural Language Processing

So far, the data in the df has been populated from 2 sources, the Watson API and the IMDB API. There will be a 3rd source of data, that will be generated by applying NLP tools to the scripts and extracting valuable statistics from them, this will be achieve with the NLTK library.

- **No. of Words:** It is the total number of words in the script, it will be a measure of the length of the script.
- **Diversity:** It is the number of unique words / the total number of words, this gives a measure of the diversity of language used.
- **Parts of speech:** Each part of speech is counted from the script, this is done with the NLTK library as shown below.

We start by tokenizing the scripts, that is, separating each word in the string as a separate one, then proceed to make calculations and gaining insights from the texts.

The function **'words'** is defined to generate this values.

The function **'similarity'** is also defined, this calculates de cosines of 2 lists to understand how alike they are. It is frequently used for recommendation systems

```python
from scipy.spatial.distance import cosine
import nltk

def similarity(list1,list2):                             
    return 1.-cosine(list1,list2) 

def words(script):
    tokens    = nltk.word_tokenize(script)
    nwords    = len(tokens)
    diversity = len(set(tokens))/float(nwords)
    tagger    = nltk.pos_tag(tokens,tagset='universal')
    
    wordL     = 0.
    for i in tokens:
        wordL += len(i)
    wordL     = wordL/nwords
    
    speech    = ['VERB','NOUN','ADP','.','ADJ','ADV','CONJ','PRON','PRT','NUM','X']
    counter   = [0.]*len(speech)
    for i in tagger:
        for j in range(len(speech)):
            if i[1] == speech[j]:
                counter[j] += 1.
    
    return [nwords,diversity,wordL,counter[0]/nwords,counter[1]/nwords,counter[2]/nwords,
            counter[3]/nwords,counter[4]/nwords,counter[5]/nwords,counter[6]/nwords,
            counter[7]/nwords,counter[8]/nwords,counter[9]/nwords,counter[10]/nwords]
```

Some changes were made here, adding new information to the NLP extracted. Doesn't change the previous results, normalized versions of the totals of parts of speech are returned by the function. All previous data is stored in the scrapped_old and data_old folders.

Parts of speech added are: **Verb, Noun, Adp, Punctuation, Adjectives, Adverbs, Conjunctions, Pronouns, Particles, Numbers and Not Recognized.**

Following is the function **'swap'** which changes the position of two columns.

```python
def swap(df,str1,str2):
    columns = list(df.columns)
    num = [columns.index(str1),columns.index(str2)]
    return df[columns[:num[0]]+[columns[num[1]]]+columns[num[0]+1:num[1]]+[columns[num[0]]]+columns[num[1]+1:]]
```

It is important to search for outliers, any indication of corrupted or wrong or statistically improbable data should be reviewed. Start by defining the outliers, will use the **'IQR'** function for such purpose.

```python
def getIQR(series,n1=.25,n2=.75,mult=1.5):
    q1 = series.quantile(n1)
    q3 = series.quantile(n2)
    dist = (q3-q1)*mult
    iqr = (series < q1-dist) | (series > q3+dist)
    return iqr
```

## Supervised Learning, Classification


Proceed to prepare data, first by splitting the data into a training set and a test set, target should also be divided accordingly. Division should be done at random to avoid biases.

Define function **'train_test'** to split the data, it provides the index of the rows that will be randomly chosen.

```python
def train_test(x,size=0.7,boot=False,seed=0):
    np.random.seed(seed)
    r = range(len(x))
    train = sorted(np.random.choice(r,size= int(len(x)*size),replace=boot))
    test = []
    for i in r:
        if i not in train:
            test.append(i)
    return train, test
```

### Tuning

**KNN Tuning**

Start with KNN, the following function 'knn_tune' takes the dataset, and performs KNR while changing the tuning parameters, accepts size of training/test sets, number of columns (features) to use, and number K for changing the neighbors. Outputs a matrix of weighted parameters vs number K.

```python
df   = pd.read_csv('data/dfcPP.csv',index_col=0)
imdb = pd.read_csv('data/imdbP.csv',index_col=0)
```

```python
def knn_tune(df,imdb,n=0.7,ind=30,start=(1,31)):
    predictions = pd.DataFrame(data=np.zeros(shape=(ind,2)),
                               columns=['uniform','distance'],index=range(start[0],start[1]))
    for seed in range(ind):
        train_index, test_index = train_test(imdb['Genre1'],n,seed)
        train        = df.iloc[train_index]
        test         = df.iloc[test_index]
        target_train = imdb['Genre1'].iloc[train_index]
        target_test  = imdb['Genre1'].iloc[test_index]
        #target_test2 = imdb['Genre1'].iloc[test_index]

        for neig in range(start[0],start[1]):
            for opt in ['uniform','distance']:
                clf = KNC(n_neighbors=neig,weights=opt)
                clf.fit(train,target_train.ravel())
                kncpred = clf.predict(test)
                count = [0.,0.]
                for i, j in zip(kncpred,target_test):
                    if i == j:
                        count[0] += 1
                    count[1] += 1
                # for i, j, k in zip(kncpred,target_test,target_test2):
                #     if i == j or i == k:
                #        count[0] += 1
                #     count[1] += 1
                predictions[opt][neig] += (count[0]/count[1])/ind
    return predictions
```

- Data = ‘Preprocessed’ or ‘Unprocessed’
    - Random seeds = 30
        - K = 30
            - Weights = ‘uniform’ or ‘distance’

Check KNN outcome for Preprocessed data, 30 random batches, 70% train, 30% test uniform & distance.

```python
knnTune = knn_tune(df,imdb)
knnTune.plot()
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('Preprocessed data')
plt.show()
```
