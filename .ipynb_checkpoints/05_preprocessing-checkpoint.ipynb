## Preprocessing

```python
import pandas as pd

df = pd.read_csv('data/df.csv',index_col=0)
```

```python
df.index = df['Script']
df.drop('Script',axis=1,inplace=True)
df.index.name = None
```

```python
from sklearn import preprocessing

x = df[df.columns[0:44]].values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df[df.columns[0:44]] = x_scaled
df.describe()
```

```python
df.to_csv('data/dfminmax.csv',encoding='utf-8')
```

```python
df[df.columns[:44]] = preprocessing.scale(df[df.columns[:44]])
df[df.columns[:44]].describe()
```

The numeric values obtianed created with the NLTK library and Watson's API have different ranges of values, will preprocess and scale them to have mean = 0 and std = 1. This is important to more correctly understand the structure of the data and for some machine learning algorithms to work correctly.

Only the numeric features are selected.

As can be seen in the previous table, the std is very close to 1 and the mean very close to 0 for all the features.
