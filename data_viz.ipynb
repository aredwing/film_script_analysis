Check for missing, 'NAs' values, define **'checkNAS'.** The information will be relevant later on. 

As it is seen, all NA's are from the IMDB data.

```python
def checkNAS(df):
    NAS = []
    for i in df.columns:
        NAS.append([sum(pd.isnull(df[i])),i])
    NAS = sorted(NAS,reverse=True)
    for i in NAS:
        if i[0] !=0:
            print i
```

```python
NAS = checkNAS(df)
NAS
```

Continue by searching for outliers, define **'getIQR'** to check which values are outliers.

```python
def getIQR(series,n1=.25,n2=.75,mult=1.5):
    q1 = series.quantile(n1)
    q3 = series.quantile(n2)
    dist = (q3-q1)*mult
    iqr = (series < q1-dist) | (series > q3+dist)
    return iqr
```

```python
IQR = {}
for feat in df.columns[:44]:
    IQR.update({feat:df[feat].ix[getIQR(df[feat])]})

IQR['WORDS'][:5]
```

### Preprocessing

The numeric values obtianed through NLP and Watson have different ranges of values, will preprocess and scale them to have mean = 0 and std = 1. Copy the current dataframe and store the preprocessed one.

```python
from copy import deepcopy
from sklearn import preprocessing

dfpp = deepcopy(df)

dfpp[dfpp.columns[:44]] = preprocessing.scale(dfpp[dfpp.columns[:44]])
dfpp[dfpp.columns[:44]].describe()
```

A quick view of the most important features in a random forest run:

```python
from sklearn.ensemble import RandomForestRegressor as RFR
nas = ~dfpp['imdbRating'].isnull()

clf = RFR(n_estimators=1000,random_state=0)
clf.fit(dfpp[dfpp.columns[:44]].ix[nas],pd.Series(dfpp['imdbRating'].ix[nas]).ravel())
```

```python
imp  = []
for i in range(len(dfpp[dfpp.columns[:44]].columns)):
    imp.append([clf.feature_importances_[i],dfpp.columns[i]])
imp = pd.DataFrame(sorted(imp,reverse=True))
imp.index = list(imp[1])
imp = imp.drop(1,1)
imp.columns = ['importance']
imp.transpose()
```

A matrix scatterplot of the top importance features seen.

```python
n = 7

pd.scatter_matrix(dfpp[list(imp.index[:n])],alpha=0.5,figsize=(10,10),diagonal='kde')
plt.show()
```

### PCA and LDA

PCA or principal component analysis can help us discover underlying factors on the evident features, associations between factors that may come from common roots, by looking at this, it is possible to associate groups of features that commonly act together.

```python
from sklearn.decomposition import PCA
import rs

columns = imp.transpose().columns[:n]
pca     = PCA(n_components=n,whiten=True)
pca.fit(dfpp[columns])

pca_results = rs.pca_results(dfpp[columns],pca)
```

Proceed to plot a correlation matrix, this is a numerical measure of how the features variate with each other. The next one, the correlation matrix plot will make it even clearer.

```python
import seaborn as sns
import numpy as np

columns = imp.transpose().columns[:n]
corr = dfpp[columns].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask, 1)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, square=True, annot=True, cmap='RdBu', fmt='+.1f')
    plt.xticks(rotation=45, ha='right');
corr = dfpp.corr()
```

Sympathy and Intellect are highly correlated, that is nice.

But what do we care about the relationship in the features, it turns out it gives us clues about parameters we can not see directly in the data, like writting style, authors, genre, etc. of the scripts.

Use the first genre listed as the identifier for each point, choose from color palette to identify each genre.

```python
n = 2
nas            = ~dfpp['Genre1'].isnull()
X              = dfpp[imp.transpose().columns[:44]].ix[nas]
targets        = pd.Series(dfpp['Genre1'].ix[nas]).ravel()

color_list = []
for name, hex in matplotlib.colors.cnames.iteritems():
    color_list.append([name,hex])

colors = []
for i in range(len(pd.unique(targets))):
    colors.append(color_list[i][0])
```

Create the PCA and plot them but first check the most common genres, the top found follow.

```python
movie_gen = []
for i in pd.unique(targets):
    movie_gen.append([sum(targets==i),i])
    
n = 8
for i in sorted(movie_gen,reverse=True)[:n]:
    print i
```

```python
pca          = PCA(n_components=n)
X_r          = pca.fit_transform(X)
lw           = 2
n = 4

target_names = ['Comedy','Drama','Action','Crime','Horror','Animation'][:n]

for color, i, target_name in zip(colors[:n],range(n),target_names):
    plt.scatter(X_r[target_name==targets,0],X_r[target_name==targets,1],color=color,alpha=.5,lw=lw,label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('PCA of Scripts dataset')
plt.show()
```

Not much to see here, the X axis is PCA1 explains the maximum variance, PCA2 or y axis explaints the second maximum.
Perform LDA to see if it provides better information of the structure of the data.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

y = []

for i in targets:
    y.append(list(pd.unique(targets)).index(i))
lda  = LDA(n_components=2)
X_r2 = lda.fit(X, y).transform(X)

for color, i, target_name in zip(colors[:n],range(n),target_names):
    plt.scatter(X_r2[target_name==targets,0],X_r2[target_name==targets,1],color=color,alpha=.5,lw=lw,label=target_name)
    
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of Scripts dataset')
plt.xlabel('LDA1')
plt.ylabel('LDA2')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

lda  = LDA(n_components=3)
X_r3 = lda.fit(X, y).transform(X)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for color, i, target_name , m in zip(colors[:n],range(n),target_names,['v','^','o','*']):
    ax.scatter(X_r3[target_name==targets,0],X_r3[target_name==targets,1],X_r3[target_name==targets,2],c=color,marker=m)

ax.set_xlabel('LDA 1')
ax.set_ylabel('LDA 2')
ax.set_zlabel('LDA 3')
ax.view_init(30,30)
plt.draw()
plt.show()
```

### Recommendation Dataset

Before going further, create a new dataset that will contain the recommendation matrix for a script (film) based on its script. To achieve such task the **'similarity'** function is define, it calculates the cosines distance from a script to another, then calculate it and save it.

```python
from scipy.spatial.distance import cosine

def similarity(list1,list2):                             
    return 1.-cosine(list1,list2) 
```

```python
side = len(df.index)

rec  = np.zeros(shape=(side,side))

for i in range(side):
    for j in range(i+1,side):
        rec[i][j] = similarity(list(df[df.columns[:30]].iloc[i]),list(df[df.columns[:30]].iloc[j]))
        rec[j][i] = rec[i][j]
```

```python
rec = pd.DataFrame(data=rec,index=df.index,columns=df.index)
rec.to_csv('data/rec.csv',encoding='utf-8')
```

```python
recs = {}
n = 5

for col in rec.columns:
    temp   = rec[col].sort_values(ascending=False).head(n)
    values = []
    for i in range(len(temp)):
        values.append([temp.index[i],temp.values[i]])
    recs.update({col:values})
```
